# STUDI KASUS 

```python

Data = [
    {"Nama": "Entong" ,
    "Jml_Tabungan" : 12000,
    "Usia" : 16
    },
    {
    "Nama": "Eneng",
    "Jml_Tabungan" : 10000,
    "Usia" : 17
    },
    {
    "Nama": "Bobby",
    "Jml_Tabungan" : 9000,
    "Usia" : 25
    },
    {
    "Nama": "Amir",
    "Jml_Tabungan" : 15000,
    "Usia" : 26
    },
    {
    "Nama": "Rendi",
    "Jml_Tabungan" : 9900,
    "Usia" : 23
    }
]
lolos = 0
tdk_lolos = 0
Nama_Lolos = []
Nama_Tidak_Lolos = []
for i in Data:
    if i["Jml_Tabungan"]>=10000:
        if 17 <= i ["Usia"] <=25:
            print(i["Nama"], end = "-> ")
            print("LOLOS")
            Nama_Lolos.append(i["Nama"])
            lolos += 1
        else:
            print(i["Nama"], end = "->")
            print("TIDAK LOLOS")
            print("\tAlasan = Usia tidak sesuai kriteria")
            Nama_Tidak_Lolos.append(i["Nama"])
            tdk_lolos += 1
    else:
        print(i["Nama"],end = "-> ")
        print("TIDAK LOLOS")
        print("\tAlasan = Jumlah tabungan tidak sesuai kriteria")
        Nama_Tidak_Lolos.append(i["Nama"])
        tdk_lolos += 1
        
print("\n-------------------RANGKUMAN-------------------")
print("Nama yang lolos\t\t= ",Nama_Lolos)
print("Nama yang tidak lolos\t= ",Nama_Tidak_Lolos)
print("\nJumlah lolos\t\t= ",lolos)
print("Jumlah Tidak Lolos\t= ",tdk_lolos)
```

# OPERATOR

OPERATOR ARITMATIKA

```python
x = 7
y = 5

hasil_penjumlahan = x + y
hasil_pengurangan = x - y
hasil_perkalian = x * y
hasil_pembagian = x / y
hasil_modulus = x % y
hasil_floor_division = x // y
hasil_x_pangkat_y = x ** y

print("Hasil x + y =", hasil_penjumlahan)
print("Hasil x - y =", hasil_pengurangan)
print("Hasil x * y =", hasil_perkalian)
print("Hasil x / y =", hasil_pembagian)
print("Hasil x % y =", hasil_modulus)
print("Hasil x // y =", hasil_floor_division)
print("Hasil x ** y =", hasil_x_pangkat_y)
```

OPERATOR PERBANDINGAN

```python
print("x > y =", x > y)
print("x < y =", x < y)
print("x >= y =", x >= y)
print("x <= y =", x <= y)
print("x != y =", x != y)
print("x == y =", x == y)
```

OPERATOR LOGIKA

```python
a = True
b = False
print("a and b = ", a and b)
print("a or b = ", a or b)
print("not(a) =", not(a))
```

# PERCABANGAN (Conditional Statement)

if...else

```python
nilai = 71
if nilai > 70:
    print("Selamat, kamu lulus")
else: # nilai_a <= 70
    print("Yah, kamu harus mengulang")
```

if...elif...else

```python
if nilai >= 90:
    print("Dapat nilai A")
elif nilai >= 80:
    print("Dapat nilai B")
elif nilai > 70:
    print("Dapat nilai C")
else:
    print("Dapat nilai D. Yah kamu harus mengulang")
```

# PERULANGAN (Loop)

```python
for i in range(4):
    print("Hello world! (iterasi ke-", i, ")")
```

```python
for i in range(1,4):
    print("Hello world! (iterasi ke-", i, ")")
```

```python
fruits = ["Mangga", "Melon", "Pisang", "Apel", "Salak"]
for fruit in fruits:
    print("Saya suka makan", fruit)
```

while loop

```python
from random import randint
j = randint(0,10)
while j < 10:
    print("Hello friend (iterasi ke-", j, ")")
    j += 1
```
