Hidrokarbon

```python
class hidrokarbon:
    def __init__(self, nama, jenis, ikatan):
        self.nama = nama
        self.jenis = jenis
        self.ikatan = ikatan
    
    def tungga(self):
        tunggal = input("Senyawa ini ikatan tunggal (y/n):")
        if tunggal.upper() == "Y" :
            print("senyawa ini ikatan tunggal")
        elif tunggal.upper() == "N" :
            print("senyawa ini ikatan rangkap")

    def info(self):
        print(f"Nama\t: {self.nama}\nJenis\t: {self.jenis}\nIkatan\t: {ikatan}")
        
        
#nama hidrokarbon
class hidrokarbon_metana(hidrokarbon):
    def __init__(self, nama, jenis, ikatan):
        super().__init__(nama, jenis, ikatan)
    
    def tunggal(self):
        tunggal = input("Senyawa ini ikatan tunggal (y/n):")
        if tunggal.upper() == "Y" :
            print("senyawa ini ikatan tunggal")
        elif tunggal.upper() == "N" :
            print("senyawa ini ikatan rangkap")
    
    def karbon(self, atom):
        for x in range(atom) :
            print("karbon karbon karbon karbon")
    
    def info_hidrokarbon_metana(self):
        print(f"Nama\t: {self.nama}\nJenis\t: {self.jenis}\nIkatan\t: {self.ikatan}")
    
class hidrokarbon_etana(hidrokarbon):
    def __init__(self, nama, jenis, ikatan):
        super().__init__(nama, jenis, ikatan)
    
    def tunggal(self):
        tunggal = input("Senyawa ini ikatan tunggal (y/n):")
        if tunggal.upper() == "Y" :
            print("senyawa ini ikatan tunggal")
        elif tunggal.upper() == "N" :
            print("senyawa ini ikatan rangkap")
    
    def karbon(self, atom):
        for x in range(atom) :
            print("karbon karbon karbon karbon")
    
    def info_hidrokarbon_etana(self):
        print(f"Nama\t: {self.nama}\nJenis\t: {self.jenis}\nIkatan\t: {self.ikatan}")
        
class hidrokarbon_propana(hidrokarbon):
    def __init__(self, nama, jenis, ikatan):
        super().__init__(nama, jenis, ikatan)
    
    def tunggal(self):
        tunggal = input("Senyawa ini ikatan tunggal (y/n):")
        if tunggal.upper() == "Y" :
            print("senyawa ini ikatan tunggal")
        elif tunggal.upper() == "N" :
            print("senyawa ini ikatan rangkap")
    
    def karbon(self, atom):
        for x in range(atom) :
            print("karbon karbon karbon karbon")
    
    def info_hidrokarbon_propana(self):
        print(f"Nama\t: {self.nama}\nJenis\t: {self.jenis}\nIkatan\t: {self.ikatan}")
        
Karbon_tetrahidrida = hidrokarbon_metana("Karbon_tetrahidrida", "alkana", "tunggal")
Karbon_tetrahidrida.info_hidrokarbon_metana()
Karbon_tetrahidrida.karbon(5)
        
```

```python
class AlatMusik:
    def __init__(self, jenis, merek, usia, warna):
        self.jenis = jenis
        self.merek = merek
        self.usia = usia
        self.warna = warna
    
    def bunyi(self, laju):
        print("bunyi alat musik ini :", bunyi)
        
    def info(self):
        print(f"Jenis\t: {self.jenis}\nmerek\t: {self.merek}\nusia\t: {self.usia}\nwarna\t: {self.warna}")


class gitar(AlatMusik):
    def __init__(self, jenis, merek, usia, warna, harga):
        super().__init__(jenis, merek, usia, warna)
        self.harga = harga
    
    def bunyi_gitar(self):
        print("jreng jreng jreng jreng")
    
    def info_gitar(self):
        print(f"Jenis\t: {self.jenis}\nmerek\t: {self.merek}\nusia\t: {self.usia}\nwarna\t: {self.warna}\nharga\t: {self.harga}")
        
            
class drum(AlatMusik):
    def __init__(self, jenis, merek, usia, warna):
        super().__init__(jenis, merek, usia, warna)
        self.harga = harga
    
    def bunyi_drum(self):
        print("drum drum drum drum")
     
    def info_drum(self):
        print(f"Jenis\t: {self.jenis}\nmerek\t: {self.merek}\nusia\t: {self.usia}\nwarna\t: {self.warna}")               
      

class piano(AlatMusik):
    def __init__(self, jenis, merek, usia, warna):
        super().__init__(jenis, merek, usia, warna)
        self.harga = harga
    
    def bunyi_piano(self):
        print("tuut tuut tuut tuut")
     
    def info_piano(self):
        print(f"Jenis\t: {self.jenis}\nmerek\t: {self.merek}\nusia\t: {self.usia}\nwarna\t: {self.warna}")   
    
    
    

gitar1 = gitar("Gitar", "Yamaha", 1, "Coklat", "Rp.1.500.000")

gitar1.info_gitar()
gitar1.bunyi_gitar()
    
    
```
