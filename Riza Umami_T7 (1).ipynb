```python
# Melakukan Coding disini
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

**LANGKAH 3**

**MENYIAPKAN DATASET**

https://raw.githubusercontent.com/jumadi-cloud/Fundamental-Python/main/Dataset/konsumen2.csv

image.png

```python
# Melakukan Coding disini
Dataset = pd.read_csv('https://raw.githubusercontent.com/jumadi-cloud/Fundamental-Python/main/Dataset/konsumen2.csv')
Dataset.keys()
```

image.png

```python
# Melakukan Coding disini
dataku = pd.DataFrame(Dataset)
dataku.head()
```

image.png

```python
# Melakukan Coding disini
X = np.asarray(Dataset)
print(X)
```

image.png

```python
# Melakukan Coding disini
plt.scatter(X[:,0],X[:,1], label='True Position')
plt.xlabel('Gaji')
plt.ylabel('Pengeluaran')
plt.title('Grafik Knosumen')
plt.show()
```

**LANGKAH 4**

**MENGGUNAKAN LIBRARY K-MEANS**

image.png

```python
# Melakukan Coding disini
kmeans = KMeans(n_clusters=2, random_state=4)
kmeans.fit(X)
```

image.png

```python
# Melakukan Coding disini
print(kmeans.cluster_centers_)
```

image.png

image.png

```python
# Melakukan Coding disini
print(kmeans.labels_)
```

image.png

```python
# Melakukan Coding disini
metrics.silhouette_score(X, kmeans.labels_, metric='euclidean')
```

**LANGKAH 5**

**MENAMPILKAN OUTPUT**

image.png

```python
# Melakukan Coding disini
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel('Gaji')
plt.ylabel('Pengeluaran')
plt.title('Grafik Knosumen')
plt.show()
```

image.png

```python
# Melakukan Coding disini
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color='black')
plt.xlabel('Gaji')
plt.ylabel('Pengeluaran')
plt.title('Grafik Knosumen')
plt.show()
```

# Melakukan Pencarian K Terbaik dengan silhouete Method

image.png

```python
# Tulis code disini
ks=10

sil_score=np.zeros(ks)
for k in range(1, ks+1):
  kmeans= KMeans(n_clusters=k, random_state=4)
  kmeans.fit(X)

  if k!=1:
    s_score= metrics.silhouette_score(X, kmeans.labels_, metric='euclidean')
    sil_score[k-1]= s_score

sil_score    

```

image.png

```python
#Code disini
plt.plot(range(1, ks+1), sil_score, 'g')
plt.legend(('silhouete score'))
plt.ylabel(('silhouete score'))
plt.xlabel('Number of Nabors (K)')
plt.tight_layout()
plt.show()
```
